# Data Visualization

Visualize IMU data for a crash detection system for a drone that is blind on it left and right sides.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### Import Dataset

```python
import os

# Join data path
current_dir = os.path.curdir
print(current_dir)
datapath = os.path.join(current_dir, "data")
print(datapath)
```

#### Extract Data
Make sure to install openpyxl library
conda to install openpyxl

```python
accel_data = pd.read_excel(r'.\data\acceleration.xlsx')
accel = np.array(accel_data)
```

```python
accel
```

```python
X = np.array(accel[:,1],dtype=np.float32)
Y = np.array(accel[:,2],dtype=np.float32)
Z = np.array(accel[:,3],dtype=np.float32)
t = np.arange(0,len(X))
```

```python
X
```

```python
plt.clf()
plt.figure(1)
plt.plot(t,X,color='red',label='X')
plt.plot(t,Y,color='blue',label="Y")
plt.plot(t,Z,color='green',label="Z")
plt.legend(['X','Y','Z'])
plt.xlabel('Time [sec]')
plt.ylabel('Acceleration [m/s^2]')
plt.title('Accelerations')
plt.show()
```
