# Data Visualization

Visualize IMU data for a crash detection system for a drone that is blind on it left and right sides.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### Import Dataset

```python
import os

# Join data path
current_dir = os.path.curdir
print(current_dir)
datapath = os.path.join(current_dir, "data")
print(datapath)
```

#### Extract Data
Make sure to install openpyxl library
conda to install openpyxl

```python
accel_data = pd.read_excel(r'.\data\acceleration.xlsx')
accel = np.array(accel_data)
orient_data = pd.read_excel(r'.\data\orientation.xlsx')
orient = np.array(orient_data)
```

```python
orient
```

```python
Xa = np.array(accel[:,1],dtype=np.float32)
Ya = np.array(accel[:,2],dtype=np.float32)
Za = np.array(accel[:,3],dtype=np.float32)
Xo = np.array(orient[:,1],dtype=np.float32)
Yo = np.array(orient[:,2],dtype=np.float32)
Zo = np.array(orient[:,3],dtype=np.float32)
t = np.arange(0,len(Xa))
```

```python
Xa
```

### Plot Accelerations

```python
plt.clf()
plt.figure(1)
plt.plot(t,Xa,color='red',label='X')
plt.plot(t,Ya,color='blue',label="Y")
plt.plot(t,Za,color='green',label="Z")
plt.legend(['X','Y','Z'])
plt.xlabel('Time [sec]')
plt.ylabel('Acceleration [m/s^2]')
plt.title('Accelerations')
plt.show()
```

```python
plt.clf()
plt.figure(1)
plt.plot(t,Za,color='green',label='Z')
plt.legend(['Z'])
plt.xlabel('Time [sec]')
plt.ylabel('Acceleration [m/s^2]')
plt.title('Accelerations')
plt.show()
```

### Plot Orientation

```python
plt.clf()
plt.figure(1)
plt.plot(t,Xo,color='red',label='X')
plt.plot(t,Yo,color='blue',label="Y")
plt.plot(t,Zo,color='green',label="Z")
plt.legend(['X','Y','Z'])
plt.xlabel('Time [sec]')
plt.ylabel('Orientation [deg]')
plt.title('Orientations')
plt.show()
```
